```python
import pandas as pd
import data_manager as dm
import numpy as np
import matplotlib.patches as mpatches

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.rcParams.update({'font.size':15})
```

```python
df = dm.load_data("setup_culindric_porosities.csv")
df['date'] =  df.date.str.replace("_", "/")
df['date'] = pd.to_datetime(df.date)
df
```

```python
df["mean"] = pd.to_numeric(df["mean"])
df["std"] = pd.to_numeric(df["std"])
df
```

```python
dm.save_dataframe(df, "porosities.xlsx")
```

```python

polymer_type = "PDLG-5002"
```

```python
def plot(polymer_type, df=df):
    df = df[df.polimer_type==polymer_type]

    fig, ax = plt.subplots(figsize=(7, 7), constrained_layout=True)

    colors = ["blue", "black", "red"]
    markers = ["o", "^", "x"]
    labels = []
    for sample_number, color, marker in zip(np.flip(df.sample_number.unique()), colors, markers):
        df_for_1_sample = df[df.sample_number==sample_number]
        s = 3000*np.pi*(df_for_1_sample["std"].to_numpy()**2)
        labels.append(mpatches.Patch(color=color, label=polymer_type+f" образец {sample_number}"))
        df_for_1_sample.plot(kind="scatter",
                             x="date",
                             y="mean",
                             ax=ax,
                             c=color,
                             s=s,
                             marker=marker
                             )
    
    ax.set_ylabel("средняя пористость")
    ax.set_xlabel("дата измерений")
    ax.set_ylim(bottom=0, top=1)
    ax.legend(loc=2, handles=labels)
    plt.xticks(rotation='vertical')
```

```python
plot("PDLG-5002")
```

```python
plot("PDL-05")
```

# Синхротрон

```python
df =  pd.DataFrame(columns = ['polimer_type', 'sample_type', 'date', 'mean', 'std'])
polimers = ["PDL-05"] * 6
sample_types = ["центр", "поверхность", "центр", "поверхность", "центр", "поверхность"]
dates = ["исходный", "исходный", "неделя 1", "неделя 1", "неделя 2", "неделя 2"]
means = [0.78, 0.69, 0.77, 0.70, 0.77, 0.68]
stds = [0.12, 0.13, 0.12, 0.14, 0.13, 0.11]

for polimer, sample_type, date, mean, std in zip(polimers, sample_types, dates, means, stds):
    df = df.append({'polimer_type': polimer, 
                    'sample_type': sample_type,
                    'date': date,
                    'mean': mean,
                    'std': std}, ignore_index=True)
df
```

```python
fig, ax = plt.subplots(figsize=(7, 7), constrained_layout=True)

colors = ["blue", "red"]
markers = ["o", "^", "x"]
labels = []
for polymer_type, sample_type, color, marker in zip(df.polimer_type, np.flip(df.sample_type.unique()), colors, markers):
    df_for_1_sample = df[df.sample_type==sample_type]
    s = 3000*np.pi*(df_for_1_sample["std"].to_numpy()**2)
    labels.append(mpatches.Patch(color=color, label=polymer_type+f" образец {sample_type}"))
    df_for_1_sample.plot(kind="scatter",
                         x="date",
                         y="mean",
                         ax=ax,
                         c=color,
                         s=40,
                         yerr = df_for_1_sample["std"].to_numpy(),
                         marker=marker,
                         alpha=0.5
                         )

ax.set_ylabel("средняя пористость")
ax.set_xlabel("дата измерений")
ax.legend(loc=8, handles=labels)
ax.set_ylim(bottom=0, top=1)
plt.xticks(rotation='vertical')
```
